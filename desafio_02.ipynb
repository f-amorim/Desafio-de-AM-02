```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
penguins = pd.read_csv('./penguins.csv')
penguins.info()
```

<h1>Tratando dados</h1>

```python
penguins.isna().sum()
```

```python
penguins = penguins.dropna()
penguins.isna().sum()
```

<h1>One Hot Encoding</h1>

```python
penguins_dt = pd.merge(penguins, pd.get_dummies(penguins, columns=['Ilha', 'sexo']))
penguins_dt.head()
```

```python
penguins_dt.columns
```

```python
X_columns = ['largura do bico', 'profundidade do bico',
       'profundidade do bico.1', 'massa corporal', 'Ilha_Biscoe',
       'Ilha_Dream', 'Ilha_Torgersen', 'sexo_female', 'sexo_male']
Y_columns = ['Espece']
```

```python
X = penguins_dt[X_columns]
Y = penguins_dt[Y_columns]
```

<h1>Split Train Test Data</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y, test_size=0.2, random_state=101)
```

<h1>Criando Modelo</h1>

```python
model = RandomForestClassifier(n_estimators=1000, random_state=101)
model.fit(x_train,y_train)
```

<h1>avaliando o modelo</h1>

```python
preds = model.predict(x_test)
preds
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("acurácia do modelo de :", accuracy_score(y_test,preds))
```
